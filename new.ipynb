```python
import re
import pandas as pd

df = pd.read_csv("vacatures.csv")

df.head()
```

```python
import matplotlib.pyplot as plt

top_titles = df['title'].value_counts().head(10)

plt.figure(figsize=(10, 6))
top_titles.plot(kind='barh', color='blue')
plt.gca().invert_yaxis()
plt.title("Top 10 meest voorkomende functietitels")
plt.xlabel("Aantal vacatures")
plt.ylabel("Functietitel")
plt.grid(axis='x')
plt.show()
```

```python
def los_maken(location: str) -> str:
    if pd.isna(location):
        return ""
    match = re.match(r"^([^,]+)", location)
    return match.group(1).strip() if match else location.strip()
```
